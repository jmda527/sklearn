```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

# 1.製造亂數

```python
noise=np.random.randn(200)
x=np.linspace(-10,10,200)
y=2*x+2*noise
plt.scatter(x,y)
```

# 2.切割test/train data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=0)
```

# 3.建立model

```python
model_jmda=LinearRegression()
```

```python
model_jmda.fit(x_train.reshape(-1,1),y_train)
```

# 4.觀察fitting完後結果

```python
train_predict=model_jmda.predict(x_train.reshape(-1,1))
test_predict=model_jmda.predict(x_test.reshape(-1,1))
```

```python
plt.figure(figsize=(13,4))
plt.subplot(1,2,1)
plt.scatter(x_train,train_predict,color='r',label='predict',alpha=0.5)
plt.scatter(x_train,y_train,color='b',label='real',alpha=0.5)
plt.xlabel('x');plt.ylabel('y')
plt.title('performance of model on train data')
plt.legend()
plt.subplot(1,2,2)
plt.scatter(x_test,test_predict,color='r',label='predict',alpha=0.5)
plt.scatter(x_test,y_test,color='b',label='real',alpha=0.5)
plt.xlabel('x');plt.ylabel('y')
plt.title('performance of model on test data')
plt.legend()
```

```python
test_score=np.round(model_jmda.score(x_test.reshape(-1,1),y_test),4)
train_score=np.round(model_jmda.score(x_train.reshape(-1,1),y_train),4)
print('train_score:[{}]     test_score:[{}]'.format(train_score,test_score))
```
